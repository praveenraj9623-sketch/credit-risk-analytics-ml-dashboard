```python
import pandas as pd
from scipy.stats import ttest_ind, chi2_contingency

df = pd.read_csv("C:/Users/admin/Desktop/credit-risk-project/data/processed/credit_dashboard_data.csv")
```

### Test 1: Income difference

```python
safe_income = df[df["TARGET"] == 0]["AMT_INCOME_TOTAL"]
risk_income = df[df["TARGET"] == 1]["AMT_INCOME_TOTAL"]

t_stat, p_value = ttest_ind(safe_income, risk_income, equal_var=False)

print("Income T-Test")
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: Significant difference in income between safe and risky customers")
else:
    print("Result: No significant difference in income")
```

### Test 2 Credit amount difference

```python
safe_credit = df[df["TARGET"] == 0]["AMT_CREDIT"]
risk_credit = df[df["TARGET"] == 1]["AMT_CREDIT"]

t_stat, p_value = ttest_ind(safe_credit, risk_credit, equal_var=False)

print("Credit Amount T-Test")
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: Significant difference in credit amount between groups")
else:
    print("Result: No significant difference in credit amount")
```

### Test 3 Education vs default

```python
education_table = pd.crosstab(df["NAME_EDUCATION_TYPE"], df["TARGET"])

chi2, p_value, dof, expected = chi2_contingency(education_table)

print("Education Chi-Square Test")
print("Chi-square:", chi2)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: Education level is significantly associated with default risk")
else:
    print("Result: No significant association between education and default risk")
```

### Test 4: Occupation vs default

```python
occupation_table = pd.crosstab(df["OCCUPATION_TYPE"], df["TARGET"])

chi2, p_value, dof, expected = chi2_contingency(occupation_table)

print("Occupation Chi-Square Test")
print("Chi-square:", chi2)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: Occupation is significantly associated with default risk")
else:
    print("Result: No significant association between occupation and default risk")
```
